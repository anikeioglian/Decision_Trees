1. Build a classification tree. Random_state =2020. Training partition 0.7. stratify = y, max_depth = 6, use Entropy (8 marks)

A. Read the dataset, explore it and determine what is the dependent variable (1 mark)

B. Store the feature and target variable. (2 marks)

C. Convert all categorical variables into dummy variables (1 mark)

D. Partition the data into training and test sets (70/30) (1 mark)

E. Fit the training data to a classification tree. (3 marks)

```python
# Enter your code here
# A. Read the dataset, explore it and determine what is the dependent variable
# read dataset
df = pd.read_csv('mushrooms.csv')
print(df.head())
# explore dataset
print(df.shape)
print(df.columns.values)
# Make sure there is no missing values
print(df.isnull().sum())
# No missing variables!
# The dependent variable should be class since it determines whetherthe mushroom is poisonous or not (p is poisonous and e is edible)
```

```python
# B. Store the feature and target variable.
col = ['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor','gill-attachment', 'gill-spacing', 'gill-size', 'gill-color', 'stalk-shape',
        'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring','stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 
        'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat']
X = df[col]
y = df['class']
```

```python
# C. Convert all categorical variables into dummy variables
df1 = pd.get_dummies(X)
print(df1)
```

```python
# D. Partition the data into training and test sets (70/30)
# Build a classification tree. Random_state =2020. Training partition 0.7. stratify = y
X = df1.iloc[:, 1:]
y = df.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(df1, y, test_size=0.3, random_state=2020, stratify=y)
```

```python
# E. Fit the training data to a classification tree.
dt = DecisionTreeClassifier(max_depth = 6, criterion = "entropy", random_state=2020)
dt.fit(X_train, y_train)
```

2. Print the confusion matrix. 
Also visualize the confusion matrix using plot_confusion_matrix from sklearn.metrics (3 marks)



```python
# Enter your code here
# Print confusion matrix
y_pred = dt.predict(X_test)
cm = metrics.confusion_matrix(y_test, y_pred)
print(cm)

# Visualize the confusion matrix
metrics.plot_confusion_matrix(dt, X_test, y_test)
plt.show()
```

3. Print the accuracy on the training partition?  (2 marks)


```python
# Enter your code here
training_pred = dt.predict(X_train)
acc_training = accuracy_score(training_pred, y_train)
print("Accuracy on Training partition:",acc_training)
```

4. Print the accuracy on the test partition? (2 marks)


```python
# Enter your code here
acc_testing = accuracy_score(y_pred, y_test)
print("Accuracy on Testing partition:",acc_testing)
```

5. Show the classification tree and save it (5 marks)


```python
# Enter your code here
plt.figure(figsize = (15,15))
fn = X.columns
cn = list(map(str,dt.classes_.tolist()))
treefig = tree.plot_tree(dt, feature_names=fn, class_names=cn, filled=True)
# Save the classification tree
plt.savefig('mushrooms.png')
plt.show()
```

6. Find and print the top three most important features in your decision tree for determining toxicity. (4 marks)

```python
# Enter your code here
feature_importance =  dt.feature_importances_
features = X_train.columns
df2 = pd.DataFrame({'features': features, 'importance': feature_importance}).nlargest(3, 'importance')
print(df2)
```
